```python
import pandas as pd
import numpy as np
```

```python
def extract_risk_surface():
    print("[*] Memuat dataset transaksi dan identitas...")
    # Sesuaikan path jika direktori dataset berbeda
    df_trans = pd.read_csv('dataset/train_transaction.csv')
    df_id = pd.read_csv('dataset/train_identity.csv')
    
    print("[*] Menggabungkan (merge) data berdasarkan TransactionID...")
    df = pd.merge(df_trans, df_id, on='TransactionID', how='left')
    
    print("\n=== 1. IMBALANCE SEVERITY ===")
    fraud_counts = df['isFraud'].value_counts()
    total_trans = len(df)
    fraud_rate = (fraud_counts[1] / total_trans) * 100
    print(f"Total Transaksi: {total_trans:,}")
    print(f"Transaksi Normal (0): {fraud_counts[0]:,} ({fraud_counts[0]/total_trans*100:.2f}%)")
    print(f"Transaksi Fraud (1): {fraud_counts[1]:,} ({fraud_rate:.2f}%)")
    
    print("\n=== 2. FINANCIAL EXPOSURE (TransactionAmt) ===")
    normal_amt = df[df['isFraud'] == 0]['TransactionAmt']
    fraud_amt = df[df['isFraud'] == 1]['TransactionAmt']
    
    print("Normal - Median: ${:.2f} | 90th Pct: ${:.2f} | Max: ${:.2f}".format(
        normal_amt.median(), normal_amt.quantile(0.90), normal_amt.max()))
    print("Fraud  - Median: ${:.2f} | 90th Pct: ${:.2f} | Max: ${:.2f}".format(
        fraud_amt.median(), fraud_amt.quantile(0.90), fraud_amt.max()))
    
    print("\n=== 3. FEATURE RELIABILITY (Missing Values) ===")
    missing_pct = (df.isnull().sum() / total_trans) * 100
    
    # Klasifikasi tingkat noise
    highly_missing = missing_pct[missing_pct > 80]
    moderate_missing = missing_pct[(missing_pct > 30) & (missing_pct <= 80)]
    clean_features = missing_pct[missing_pct <= 30]
    
    print(f"Total Fitur: {len(df.columns)}")
    print(f"Fitur Handal (NaN <= 30%): {len(clean_features)} kolom")
    print(f"Fitur Abu-abu (NaN 31% - 80%): {len(moderate_missing)} kolom")
    print(f"Fitur Noise/Manipulable (NaN > 80%): {len(highly_missing)} kolom")
    
    print("\n[*] Pemetaan Risk Surface selesai.")

if __name__ == "__main__":
    extract_risk_surface()
```
