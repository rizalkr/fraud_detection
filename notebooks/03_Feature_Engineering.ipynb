# 04 Missing Value Handling

## Objective

Identify and handle missing values using strategies that preserve meaningful information while minimizing the introduction of bias into the dataset.

## Business Question

**How should missing values be treated to produce a reliable feature set for fraud prediction?**

### 4.1 Missing Value Overview

The exploratory data analysis revealed that missing values are widely distributed across the dataset. Some variables contain only a small proportion of missing observations, whereas others—particularly several identity-related attributes—contain missing values in more than 90% of the records.

Rather than removing all highly incomplete features, the missing-value treatment strategy will be determined based on both the amount of missing data and the business meaning of each feature.

```python
missing_summary = (
    df.isnull()
      .sum()
      .to_frame("Missing")
)

missing_summary["Missing (%)"] = (
    missing_summary["Missing"] / len(df) * 100
).round(2)

missing_summary = (
    missing_summary
    .query("Missing > 0")
    .sort_values("Missing (%)", ascending=False)
)

missing_summary.head(20)
```

### 4.2 Missing Value Strategy

Missing values in this dataset originate from different sources and therefore cannot be handled using a single imputation strategy. Based on the exploratory data analysis, missing values are categorized into three groups:

1. **Ordinary transaction features**, where missing values are assumed to result from incomplete observations and can be imputed using statistical methods.

2. **Identity-related features**, where missing values often indicate that no identity information was collected for the transaction. In this case, the absence of information itself may carry predictive value and should therefore be preserved whenever appropriate.

3. **Extremely sparse features**, which contain a very high proportion of missing values. These features will be evaluated individually to determine whether they should be retained, transformed, or removed, depending on both their business meaning and predictive potential.

The following sections apply these principles to construct a consistent and machine learning-ready dataset.

### 4.3 Apply Missing Value Treatment

The missing value treatment is performed according to the strategy defined in the previous section. Different feature groups are handled using different approaches based on their business meaning and data characteristics.

The treatment process follows these principles:

- Numerical transaction features are imputed using statistical values.
- Categorical features preserve missing information as an explicit category.
- Identity-related features retain missingness whenever it represents meaningful behavioral information.
- Extremely sparse features are evaluated individually before deciding whether to retain or remove them.

#### 4.3.1 Missing Value Treatment Rules

The following rules are applied during missing value treatment:

| Feature Group | Treatment | Rationale |
|---------------|-----------|-----------|
| Transaction numerical features | Median imputation | Preserves the overall distribution while reducing the influence of extreme values. |
| Transaction categorical features | `"Missing"` category | Missing values are treated as a valid transaction category. |
| Identity numerical features | `-999` | Missing identity information may represent meaningful transaction behavior rather than incomplete data. A sentinel value preserves this information for subsequent modeling. |
| Identity categorical features | `"Missing"` category | Preserves the absence of identity information as a potential predictive signal. |
| Extremely sparse features | Evaluated individually | Features with extremely high missing rates are assessed separately before deciding whether to retain or remove them. |

#### 4.3.2 Missing Value Imputation

The missing value treatment rules established in the previous section are applied to each feature group.

To ensure consistency with the exploratory data analysis, different imputation methods are used according to the characteristics of each feature group. Transaction-related numerical features are imputed using statistical values, whereas identity-related features preserve missingness as meaningful behavioral information through sentinel values or explicit categories.

The following implementation applies these rules across the dataset.

```python
# ===============================
# Identify feature groups
# ===============================

identity_columns = [
    col for col in df.columns
    if col.startswith("id_")
]

identity_numeric = [
    col for col in identity_columns
    if col in numerical_features
]

identity_categorical = [
    col for col in identity_columns
    if col in categorical_features
]

transaction_numeric = [
    col for col in numerical_features
    if col not in identity_numeric
]

transaction_categorical = [
    col for col in categorical_features
    if col not in identity_categorical
]

print(f"Transaction numerical features   : {len(transaction_numeric)}")
print(f"Transaction categorical features : {len(transaction_categorical)}")
print(f"Identity numerical features      : {len(identity_numeric)}")
print(f"Identity categorical features    : {len(identity_categorical)}")


# ===============================
# Apply imputation
# ===============================

# Transaction numerical features
for col in transaction_numeric:
    if df[col].isna().any():
        df[col] = df[col].fillna(df[col].median())

# Identity numerical features
for col in identity_numeric:
    df[col] = df[col].fillna(-999)

# Transaction categorical features
for col in transaction_categorical:
    df[col] = df[col].fillna("Missing")

# Identity categorical features
for col in identity_categorical:
    df[col] = df[col].fillna("Missing")
```

The feature groups are separated before applying imputation because transaction and identity features follow different missing value strategies.

Transaction features primarily represent transaction behavior and are treated using conventional statistical imputation, whereas identity features preserve missingness as a potentially informative behavioral signal identified during the exploratory data analysis.

#### 4.3.3 Verification

After applying the missing value treatment, the dataset is verified to ensure that the predefined strategy has been implemented successfully.

The verification focuses on confirming that:

- Missing values have been handled according to their respective feature groups.
- No unexpected missing values remain in the dataset.
- The dataset is ready for subsequent feature engineering and model development.

```python
remaining_missing = (
    df.isna()
      .sum()
      .sort_values(ascending=False)
)

remaining_missing = remaining_missing[remaining_missing > 0]

print(f"Remaining features with missing values : {len(remaining_missing)}")

remaining_missing.head(20)
```

```python
print(f"Dataset shape : {df.shape}")
print(f"Total missing values : {df.isna().sum().sum()}")
```

The verification confirms that all missing values have been treated according to the predefined strategy. No remaining missing values are found in the dataset, indicating that the feature engineering process has successfully produced a complete dataset for the next stage of model development.
