# 04 Target Variable Analysis

## Objective

Understand the prevalence of fraudulent transactions and assess the severity of the fraud problem.

## Business Question

**How severe is the fraud problem?**


```python
total_transactions = len(df_trans)

fraud_count = df_trans["isFraud"].value_counts()

legitimate_transactions = fraud_count.get(0, 0)
fraud_transactions = fraud_count.get(1, 0)

fraud_rate = fraud_transactions / total_transactions
fraud_rate_percentage = round(fraud_rate * 100, 2)

summary_df = pd.DataFrame({
    "Metric": [
        "Total Transactions",
        "Legitimate Transactions",
        "Fraud Transactions",
        "Fraud Rate",
        "Fraud Rate (%)"
    ],
    "Value": [
        total_transactions,
        legitimate_transactions,
        fraud_transactions,
        fraud_rate,
        fraud_rate_percentage
    ]
})

summary_df
```

```python
import matplotlib.pyplot as plt
from matplotlib.ticker import StrMethodFormatter

fraud_distribution = (
    df_trans["isFraud"]
    .value_counts()
    .sort_index()
)

labels = ["Legitimate", "Fraud"]

plt.figure(figsize=(7, 5))

bars = plt.bar(labels, fraud_distribution.to_list())

for bar in bars:
    height = bar.get_height()
    percentage = height / total_transactions * 100
    plt.text(
        x=bar.get_x() + bar.get_width() / 2,
        y=height,
        s=f"{int(height):,}\n({percentage: .1f}%)",
        ha="center",
        va="bottom"
    )

plt.gca().yaxis.set_major_formatter(
    StrMethodFormatter("{x:,.0f}")
)
plt.ylim(top=max(fraud_distribution) * 1.10)
plt.title("Transaction Distribution by Class")
plt.xlabel("Transaction Class")
plt.ylabel("Number of Transactions")
plt.grid(
    axis="y",
    linestyle="--",
    alpha=0.3
)
plt.show()
```

### Business Interpretation

Fraudulent transactions account for **3.50%** of all transactions in the dataset, indicating that fraud represents a minority class.

Although the dataset is imbalanced, it still contains more than **20,000 confirmed fraud cases**, making fraudulent activity operationally significant for analysis.

Compared with real-world payment systems, this fraud rate appears relatively high. However, since this dataset was collected for a fraud detection task, the observed prevalence should not be interpreted as representative of the actual fraud rate in production environments.

At this stage, the findings suggest that fraud is a meaningful business problem while reinforcing the need for a **behavior-based decision engine** rather than decisions based solely on transaction frequency.


# 05 Transaction Behavior Analysis

## Objective

Explore transaction characteristics and identify behavioral patterns associated with fraudulent activities.

## Business Question

**Which transaction behaviors indicate higher fraud risk?**

```python
df_trans["TransactionAmt"].describe()
```

```python
df_trans["TransactionAmt"].describe(
    percentiles=[0.90, 0.95, 0.99]
)
```

```python
p99_transaction_amount = df_trans["TransactionAmt"].quantile(0.99)
transaction_amount_p99 = df_trans[
    df_trans["TransactionAmt"] <= p99_transaction_amount
]
```

```python
plt.figure(figsize=(10, 5))

plt.hist(
    transaction_amount_p99["TransactionAmt"],
    bins=50,
    edgecolor="black"
)

plt.title("Transaction Amount Distribution (Up to 99th Percentile)")
plt.xlabel("Transaction Amount")
plt.ylabel("Number of Transactions")

plt.grid(axis="y", alpha=0.3)

plt.tight_layout()
plt.show()
```

```python
legitimate_transaction_amount = transaction_amount_p99.loc[
    transaction_amount_p99["isFraud"] == 0,
    "TransactionAmt"
]

fraud_transaction_amount = transaction_amount_p99.loc[
    transaction_amount_p99["isFraud"] == 1,
    "TransactionAmt"
]
```

```python
plt.figure(figsize=(10, 5))

plt.hist(
    legitimate_transaction_amount,
    bins=50,
    density=True,
    alpha=0.6,
    label="Legitimate"
)

plt.hist(
    fraud_transaction_amount,
    bins=50,
    density=True,
    alpha=0.6,
    label="Fraud"
)

plt.title("Transaction Amount Distribution by Transaction Class (Up to 99th Percentile)")
plt.xlabel("Transaction Amount")
plt.ylabel("Density")

plt.legend()
plt.grid(axis="y", alpha=0.3)

plt.tight_layout()
plt.show()
```

```python
bins=20
bin_edges = np.linspace(
    0,
    p99_transaction_amount,
    20
)
```

```python
transaction_amount_p99["amount_bin"] = pd.cut(
    transaction_amount_p99["TransactionAmt"],
    bins=bin_edges,
    include_lowest=True
)
```

```python
fraud_rate_by_bin = (
    transaction_amount_p99
    .groupby(
        pd.cut(
            transaction_amount_p99["TransactionAmt"],
            bins=bin_edges,
            include_lowest=True
        ),
        observed=True
    )
    .agg(
        total_transactions=("isFraud", "count"),
        fraud_transactions=("isFraud", "sum")
    )
)

fraud_rate_by_bin["fraud_rate_decimal"] = (
    fraud_rate_by_bin["fraud_transactions"]
    / fraud_rate_by_bin["total_transactions"]   
    
)

fraud_rate_by_bin["fraud_rate"] = (
    fraud_rate_by_bin["fraud_rate_decimal"] * 100
).round(2)

fraud_rate_by_bin
```

### Key Findings

The analysis of **Transaction Amount** reveals several important observations:

- **Fraudulent transactions are observed across all transaction amount ranges**, indicating that fraud is not limited to high-value transactions.

- **Fraud rates become more variable in higher transaction amount ranges, while transaction frequency decreases substantially**, making risk estimates less stable in those intervals.

- **Transaction Amount should be treated as one behavioral signal rather than a standalone decision criterion.** Subsequent analysis should incorporate transaction timing and identity-related attributes from the `train_identity` dataset to build a more reliable behavior-based fraud decision engine.


### Possible Explanations (Hypotheses)

The findings above describe patterns directly observed from the data. The following hypotheses provide possible explanations for these patterns and should be interpreted as assumptions for future investigation rather than conclusions supported by the current dataset.

- **Transaction distribution and fraud strategy:** Lower transaction amounts dominate overall payment activity, while fraudsters may also intentionally choose smaller transaction amounts to reduce the likelihood of detection or customer intervention.

- **Risk controls in payment systems:** Higher-value transactions may be subject to additional verification or stricter fraud prevention controls by financial institutions, potentially reducing the number of successful fraudulent transactions.

- **Business impact of high-value fraud:** High-value fraudulent transactions may occur less frequently but result in substantially greater financial losses, suggesting that transaction frequency alone may not adequately represent overall fraud severity.
