# 04 Target Variable Analysis

## Objective

Understand the prevalence of fraudulent transactions and assess the severity of the fraud problem.

## Business Question

**How severe is the fraud problem?**


```python
total_transactions = len(df_trans)

fraud_count = df_trans["isFraud"].value_counts()

legitimate_transactions = fraud_count.get(0, 0)
fraud_transactions = fraud_count.get(1, 0)

fraud_rate = fraud_transactions / total_transactions
fraud_rate_percentage = round(fraud_rate * 100, 2)

summary_df = pd.DataFrame({
    "Metric": [
        "Total Transactions",
        "Legitimate Transactions",
        "Fraud Transactions",
        "Fraud Rate",
        "Fraud Rate (%)"
    ],
    "Value": [
        total_transactions,
        legitimate_transactions,
        fraud_transactions,
        fraud_rate,
        fraud_rate_percentage
    ]
})

summary_df
```

```python
import matplotlib.pyplot as plt
from matplotlib.ticker import StrMethodFormatter

fraud_distribution = (
    df_trans["isFraud"]
    .value_counts()
    .sort_index()
)

labels = ["Legitimate", "Fraud"]

plt.figure(figsize=(7, 5))

bars = plt.bar(labels, fraud_distribution.to_list())

for bar in bars:
    height = bar.get_height()
    percentage = height / total_transactions * 100
    plt.text(
        x=bar.get_x() + bar.get_width() / 2,
        y=height,
        s=f"{int(height):,}\n({percentage: .1f}%)",
        ha="center",
        va="bottom"
    )

plt.gca().yaxis.set_major_formatter(
    StrMethodFormatter("{x:,.0f}")
)
plt.ylim(top=max(fraud_distribution) * 1.10)
plt.title("Transaction Distribution by Class")
plt.xlabel("Transaction Class")
plt.ylabel("Number of Transactions")
plt.grid(
    axis="y",
    linestyle="--",
    alpha=0.3
)
plt.show()
```

### Business Interpretation

Fraudulent transactions account for **3.50%** of all transactions in the dataset, indicating that fraud represents a minority class.

Although the dataset is imbalanced, it still contains more than **20,000 confirmed fraud cases**, making fraudulent activity operationally significant for analysis.

Compared with real-world payment systems, this fraud rate appears relatively high. However, since this dataset was collected for a fraud detection task, the observed prevalence should not be interpreted as representative of the actual fraud rate in production environments.

At this stage, the findings suggest that fraud is a meaningful business problem while reinforcing the need for a **behavior-based decision engine** rather than decisions based solely on transaction frequency.
