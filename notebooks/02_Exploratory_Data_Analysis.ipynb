# 04 Target Variable Analysis

## Objective

Understand the prevalence of fraudulent transactions and assess the severity of the fraud problem.

## Business Question

**How severe is the fraud problem?**


```python
total_transactions = len(df_trans)

fraud_count = df_trans["isFraud"].value_counts()

legitimate_transactions = fraud_count.get(0, 0)
fraud_transactions = fraud_count.get(1, 0)

fraud_rate = fraud_transactions / total_transactions
fraud_rate_percentage = round(fraud_rate * 100, 2)

summary_df = pd.DataFrame({
    "Metric": [
        "Total Transactions",
        "Legitimate Transactions",
        "Fraud Transactions",
        "Fraud Rate",
        "Fraud Rate (%)"
    ],
    "Value": [
        total_transactions,
        legitimate_transactions,
        fraud_transactions,
        fraud_rate,
        fraud_rate_percentage
    ]
})

summary_df
```

```python
import matplotlib.pyplot as plt
from matplotlib.ticker import StrMethodFormatter

fraud_distribution = (
    df_trans["isFraud"]
    .value_counts()
    .sort_index()
)

labels = ["Legitimate", "Fraud"]

plt.figure(figsize=(7, 5))

bars = plt.bar(labels, fraud_distribution.to_list())

for bar in bars:
    height = bar.get_height()
    percentage = height / total_transactions * 100
    plt.text(
        x=bar.get_x() + bar.get_width() / 2,
        y=height,
        s=f"{int(height):,}\n({percentage: .1f}%)",
        ha="center",
        va="bottom"
    )

plt.gca().yaxis.set_major_formatter(
    StrMethodFormatter("{x:,.0f}")
)
plt.ylim(top=max(fraud_distribution) * 1.10)
plt.title("Transaction Distribution by Class")
plt.xlabel("Transaction Class")
plt.ylabel("Number of Transactions")
plt.grid(
    axis="y",
    linestyle="--",
    alpha=0.3
)
plt.show()
```

### Business Interpretation

Fraudulent transactions account for **3.50%** of all transactions in the dataset, indicating that fraud represents a minority class.

Although the dataset is imbalanced, it still contains more than **20,000 confirmed fraud cases**, making fraudulent activity operationally significant for analysis.

Compared with real-world payment systems, this fraud rate appears relatively high. However, since this dataset was collected for a fraud detection task, the observed prevalence should not be interpreted as representative of the actual fraud rate in production environments.

At this stage, the findings suggest that fraud is a meaningful business problem while reinforcing the need for a **behavior-based decision engine** rather than decisions based solely on transaction frequency.


# 05 Transaction Behavior Analysis

## Objective

Explore transaction characteristics and identify behavioral patterns associated with fraudulent activities.

## Business Question

**Which transaction behaviors indicate higher fraud risk?**

```python
df_trans["TransactionAmt"].describe()
```

```python
df_trans["TransactionAmt"].describe(
    percentiles=[0.90, 0.95, 0.99]
)
```

```python
p99_transaction_amount = df_trans["TransactionAmt"].quantile(0.99)
transaction_amount_p99 = df_trans[
    df_trans["TransactionAmt"] <= p99_transaction_amount
]
```

```python
plt.figure(figsize=(10, 5))

plt.hist(
    transaction_amount_p99["TransactionAmt"],
    bins=50,
    edgecolor="black"
)

plt.title("Transaction Amount Distribution (Up to 99th Percentile)")
plt.xlabel("Transaction Amount")
plt.ylabel("Number of Transactions")

plt.grid(axis="y", alpha=0.3)

plt.tight_layout()
plt.show()
```

```python
legitimate_transaction_amount = transaction_amount_p99.loc[
    transaction_amount_p99["isFraud"] == 0,
    "TransactionAmt"
]

fraud_transaction_amount = transaction_amount_p99.loc[
    transaction_amount_p99["isFraud"] == 1,
    "TransactionAmt"
]
```

```python
plt.figure(figsize=(10, 5))

plt.hist(
    legitimate_transaction_amount,
    bins=50,
    density=True,
    alpha=0.6,
    label="Legitimate"
)

plt.hist(
    fraud_transaction_amount,
    bins=50,
    density=True,
    alpha=0.6,
    label="Fraud"
)

plt.title("Transaction Amount Distribution by Transaction Class (Up to 99th Percentile)")
plt.xlabel("Transaction Amount")
plt.ylabel("Density")

plt.legend()
plt.grid(axis="y", alpha=0.3)

plt.tight_layout()
plt.show()
```

```python
bins=20
bin_edges = np.linspace(
    0,
    p99_transaction_amount,
    20
)
```

```python
transaction_amount_p99["amount_bin"] = pd.cut(
    transaction_amount_p99["TransactionAmt"],
    bins=bin_edges,
    include_lowest=True
)
```

```python
fraud_rate_by_bin = (
    transaction_amount_p99
    .groupby(
        pd.cut(
            transaction_amount_p99["TransactionAmt"],
            bins=bin_edges,
            include_lowest=True
        ),
        observed=True
    )
    .agg(
        total_transactions=("isFraud", "count"),
        fraud_transactions=("isFraud", "sum")
    )
)

fraud_rate_by_bin["fraud_rate_decimal"] = (
    fraud_rate_by_bin["fraud_transactions"]
    / fraud_rate_by_bin["total_transactions"]   
    
)

fraud_rate_by_bin["fraud_rate"] = (
    fraud_rate_by_bin["fraud_rate_decimal"] * 100
).round(2)

fraud_rate_by_bin
```

### Key Findings

The analysis of **Transaction Amount** reveals several important observations:

- **Fraudulent transactions are observed across all transaction amount ranges**, indicating that fraud is not limited to high-value transactions.

- **Fraud rates become more variable in higher transaction amount ranges, while transaction frequency decreases substantially**, making risk estimates less stable in those intervals.

- **Transaction Amount should be treated as one behavioral signal rather than a standalone decision criterion.** Subsequent analysis should incorporate transaction timing and identity-related attributes from the `train_identity` dataset to build a more reliable behavior-based fraud decision engine.


### Possible Explanations (Hypotheses)

The findings above describe patterns directly observed from the data. The following hypotheses provide possible explanations for these patterns and should be interpreted as assumptions for future investigation rather than conclusions supported by the current dataset.

- **Transaction distribution and fraud strategy:** Lower transaction amounts dominate overall payment activity, while fraudsters may also intentionally choose smaller transaction amounts to reduce the likelihood of detection or customer intervention.

- **Risk controls in payment systems:** Higher-value transactions may be subject to additional verification or stricter fraud prevention controls by financial institutions, potentially reducing the number of successful fraudulent transactions.

- **Business impact of high-value fraud:** High-value fraudulent transactions may occur less frequently but result in substantially greater financial losses, suggesting that transaction frequency alone may not adequately represent overall fraud severity.

## 6.1.1 Transaction Timeline

### Objective

Explore transaction activity over the dataset timeline to identify temporal changes in legitimate and fraudulent transaction volumes.

### Business Question

**Does fraudulent transaction activity change over the course of the observed transaction timeline?**

```python
df_time["TransactionDT"].describe()
```

```python
seconds = df_time["TransactionDT"].max()

print(f"Hours : {seconds / 3600:.2f}")
print(f"Days  : {seconds / 86400:.2f}")
```

```python
# Number of seconds in one day
seconds_per_day = 24 * 60 * 60

# Convert relative timestamp into transaction day
df_time["transaction_day"] = (
    df_trans["TransactionDT"] // seconds_per_day
).astype(int)
```

```python
df_time["transaction_day"].describe()
```

```python
# Number of seconds in one day
seconds_per_day = 24 * 60 * 60

# Convert relative timestamp into transaction day
df_time["transaction_day"] = (
    df_time["TransactionDT"] // seconds_per_day
).astype(int)
```

```python
timeline_summary = (
    df_time
    .groupby("transaction_day")
    .agg(
        total_transactions=("isFraud", "size"),
        fraud_transactions=("isFraud", "sum")
    )
    .reset_index()
)
```

```python
timeline_summary["fraud_rate"] = (
    timeline_summary["fraud_transactions"]
    / timeline_summary["total_transactions"]
)
```

```python
# Calculate legitimate transactions per day
timeline_summary["legitimate_transactions"] = (
    timeline_summary["total_transactions"]
    - timeline_summary["fraud_transactions"]
)
```

```python
timeline_summary = timeline_summary[
    [
        "transaction_day",
        "total_transactions",
        "legitimate_transactions",
        "fraud_transactions",
        "fraud_rate",
    ]
]
```

```python
display(timeline_summary.describe())
display(timeline_summary.head())
```

```python
fig, axes = plt.subplots(
    nrows=2,
    ncols=1,
    figsize=(14, 8),
    sharex=True
)

# Legitimate Transactions
axes[0].plot(
    timeline_summary["transaction_day"],
    timeline_summary["legitimate_transactions"],
    linewidth=2,
    color="tab:blue"
)

axes[0].set_title("Daily Legitimate Transactions")
axes[0].set_ylabel("Transactions")
axes[0].grid(alpha=0.3)

# Fraud Transactions
axes[1].plot(
    timeline_summary["transaction_day"],
    timeline_summary["fraud_transactions"],
    linewidth=2,
    color="tab:orange"
)

axes[1].set_title("Daily Fraud Transactions")
axes[1].set_xlabel("Transaction Day")
axes[1].set_ylabel("Transactions")
axes[1].grid(alpha=0.3)

plt.tight_layout()
plt.show()
```

```python
plt.figure(figsize=(14, 5))

plt.plot(
    timeline_summary["transaction_day"],
    timeline_summary["fraud_rate"],
    linewidth=2,
    color="tab:red"
)

plt.title("Daily Fraud Rate Throughout the Observation Period")
plt.xlabel("Transaction Day")
plt.ylabel("Fraud Rate")

plt.grid(alpha=0.3)

plt.tight_layout()
plt.show()
```

```python
# Divide observation period into approximately 30-day segments
timeline_summary["period"] = pd.cut(
    timeline_summary["transaction_day"],
    bins=[0, 30, 60, 90, 120, 150, 182],
    labels=[
        "Day 1–30",
        "Day 31–60",
        "Day 61–90",
        "Day 91–120",
        "Day 121–150",
        "Day 151–182"
    ],
    include_lowest=True
)
```

```python
fraud_rate_period_summary = (
    timeline_summary
    .groupby("period", observed=True)
    .agg(
        average_transactions=("total_transactions", "mean"),
        average_fraud_rate=("fraud_rate", "mean"),
        median_fraud_rate=("fraud_rate", "median"),
        std_fraud_rate=("fraud_rate", "std")
    )
    .round(4)
)

fraud_rate_period_summary
```

## Key Findings

### 1. Daily fraud rate exhibited a clear baseline shift after the first month.

The first 30 days recorded the lowest average fraud rate (2.64%), despite having the highest average daily transaction volume (4,478 transactions). From Day 31 to Day 120, the average fraud rate consistently increased to approximately 4.0%, indicating a sustained change in fraud behavior rather than isolated spikes.

---

### 2. Changes in fraud rate were not directly associated with transaction volume.

Although daily transaction volume decreased substantially after the first month (from approximately 4,478 to around 3,000 transactions per day), the fraud rate increased and remained elevated. Similarly, the decline in fraud rate after Day 120 was not accompanied by a comparable reduction in transaction volume. These observations suggest that fluctuations in fraud rate cannot be explained solely by changes in transaction volume.

---

### 3. Daily fraud rate remained consistently volatile throughout the observation period.

The standard deviation of fraud rate remained relatively stable across all six observation periods (approximately 0.008–0.010). This indicates that the primary temporal change was a shift in the average fraud-rate baseline rather than an increase in day-to-day variability.

---

## Hypotheses

### H1. Temporal changes in fraud activity are likely driven by behavioral factors rather than transaction volume.

The increase in fraud rate after the first month occurred despite a significant reduction in average daily transaction volume, suggesting that the observed baseline shift may reflect changes in fraudulent behavior or operational conditions rather than increased transaction activity alone. This hypothesis should be validated using identity-related and behavioral attributes from the `train_identity` dataset.

---

### H2. Fraudsters may exhibit an adaptation phase before increasing attack intensity.

The transition from a low-fraud period to a sustained higher fraud-rate period may indicate an adaptation phase in fraudulent activity, where attackers initially operate conservatively before increasing activity. Validation requires temporal analysis of identity attributes, device information, and behavioral patterns.

---

### H3. Temporal information is likely to provide predictive value for fraud detection.

Because fraud-rate baseline changes over time while transaction volume remains relatively stable after the first month, temporal features may contain predictive signals beyond simple transaction chronology. Subsequent feature engineering should evaluate whether temporal attributes improve fraud-risk prediction when combined with identity and behavioral features.

## 6.1.2 Time-of-Day Analysis
### Objective

Analyze how fraud activity varies throughout the day to identify whether fraudulent transactions concentrate during particular hours.

### Business Question

**Do fraudulent transactions occur more frequently during specific hours of the day compared with legitimate transactions?**

```python
SECONDS_PER_DAY = 24 * 60 * 60
SECONDS_PER_HOUR = 60 * 60

df_time["transaction_hour"] = (
    (df_time["TransactionDT"] % SECONDS_PER_DAY) // SECONDS_PER_HOUR
).astype(int)
```

```python
display(df_time["transaction_hour"].describe())
display(df_time["transaction_hour"].value_counts().sort_index())
```

```python
hourly_summary = (
    df_time
    .groupby("transaction_hour")
    .agg(
        total_transactions=("isFraud", "size"),
        fraud_transactions=("isFraud", "sum")
    )
    .reset_index()
)

hourly_summary["legitimate_transactions"] = (
    hourly_summary["total_transactions"]
    - hourly_summary["fraud_transactions"]
)

hourly_summary.head()
```

```python
fig, axes = plt.subplots(
    nrows=2,
    ncols=1,
    figsize=(14, 8),
    sharex=True
)

# Legitimate Transactions
axes[0].plot(
    hourly_summary["transaction_hour"],
    hourly_summary["legitimate_transactions"],
    linewidth=2,
    color="tab:blue"
)

axes[0].set_title("Hourly Legitimate Transactions")
axes[0].set_ylabel("Transactions")
axes[0].grid(alpha=0.3)

# Fraud Transactions
axes[1].plot(
    hourly_summary["transaction_hour"],
    hourly_summary["fraud_transactions"],
    linewidth=2,
    color="tab:orange"
)

axes[1].set_title("Hourly Fraud Transactions")
axes[1].set_xlabel("Hour of Day")
axes[1].set_ylabel("Transactions")
axes[1].grid(alpha=0.3)

axes[1].set_xticks(range(24))
axes[1].set_xticklabels([f"{h:02d}:00" for h in range(24)])

plt.tight_layout()
plt.show()
```

## Key Findings

### 1. Legitimate and fraudulent transactions exhibit highly similar hourly patterns.

Both transaction types follow the same daily cycle, with activity decreasing during the early morning hours and increasing steadily from the afternoon until late evening. This indicates that fraud volume largely follows overall transaction volume throughout the day.

---

### 2. No obvious high-risk hour is observed based on transaction volume alone.

Although fraudulent transactions are more numerous during periods with high transaction activity, the hourly distribution closely mirrors legitimate transactions. This suggests that transaction volume alone is insufficient to identify specific high-risk hours.

---

## Hypothesis

### H1. Hour-of-day alone is unlikely to be a strong standalone predictor of fraud.

The similarity between legitimate and fraudulent hourly transaction patterns suggests that temporal information at the hourly level may provide limited discriminatory power by itself. Future analysis should evaluate hour-of-day together with transaction amount, transaction sequence, and identity-related attributes from the `train_identity` dataset to determine whether temporal context contributes additional predictive value.

## 6.2.1 ProductCD Analysis

### Objective

Analyze the distribution of transaction product categories and evaluate whether fraud occurrence differs across ProductCD groups.

### Business Question

**Do certain ProductCD categories exhibit higher fraud rates than others?**

```python
product_summary = (
    df_trans["ProductCD"]
    .value_counts(dropna=False)
    .rename_axis("ProductCD")
    .reset_index(name="total_transactions")
)

product_summary["percentage"] = (
    product_summary["total_transactions"]
    / product_summary["total_transactions"].sum()
    * 100
).round(2)

product_summary
```

```python
product_summary = (
    df_trans
    .groupby("ProductCD")
    .apply(
        lambda x: pd.Series({
            "total_transactions": len(x),
            "fraud_transactions": x["isFraud"].sum(),
            "total_transaction_amount": x["TransactionAmt"].sum(),
            "average_transaction_amount": x["TransactionAmt"].mean(),
            "fraud_transaction_amount": x.loc[
                x["isFraud"] == 1,
                "TransactionAmt"
            ].sum()
        })
    )
    .reset_index()
)

product_summary["legitimate_transactions"] = (
    product_summary["total_transactions"]
    - product_summary["fraud_transactions"]
)

product_summary["fraud_rate_decimal"] = (
    product_summary["fraud_transactions"]
    / product_summary["total_transactions"]
)

product_summary["fraud_rate"] = (
    product_summary["fraud_rate_decimal"] * 100
).round(2)

product_summary["total_transaction_amount"] = (
    product_summary["total_transaction_amount"].round(2)
)

product_summary["average_transaction_amount"] = (
    product_summary["average_transaction_amount"].round(2)
)

product_summary["fraud_transaction_amount"] = (
    product_summary["fraud_transaction_amount"].round(2)
)

product_summary = product_summary.sort_values(
    "total_transactions",
    ascending=False
)

product_summary
```

## Key Findings

### 1. Product W dominates transaction volume, while Product C contributes nearly the same number of fraudulent transactions despite a much smaller transaction volume.

Product W accounts for approximately **74%** of all transactions and records the highest number of fraudulent transactions. However, Product C contributes almost the same number of fraudulent transactions despite representing only **11.6%** of total transactions, indicating a disproportionately high concentration of fraud within this product category.

---

### 2. Product W represents the largest financial exposure, whereas Product C exhibits the highest fraud concentration.

Although Product W has the lowest fraud rate (**2.04%**), it contributes the largest total fraud amount due to its dominant transaction volume. In contrast, Product C records the highest fraud rate (**11.69%**), suggesting that different ProductCD categories represent different dimensions of fraud risk: financial exposure versus fraud likelihood.

---

### 3. Product categories exhibit distinct transaction characteristics.

Product R records the highest average transaction amount (**168.31**) while maintaining a moderate fraud rate (**3.78%**). This indicates that transaction characteristics differ substantially across ProductCD categories, suggesting that ProductCD may provide useful behavioral context beyond simple fraud frequency.

---

## Hypothesis

### H1. Product C may possess transaction characteristics that make it more attractive or vulnerable to fraudulent activity.

The disproportionately high fraud rate observed in Product C suggests that this transaction category may contain underlying characteristics associated with increased fraud risk. Further analysis using identity-related features, transaction behavior, and additional contextual attributes is required to determine the underlying cause.

---

## Analyst Note (Future Recommendation)

- Prioritize monitoring for **Product W** because it represents the largest transaction volume and the highest potential financial exposure.
- Allocate dedicated fraud detection resources to **Product C**, as it exhibits the highest concentration of fraudulent transactions relative to its transaction volume.
- Future fraud mitigation strategies should balance **fraud likelihood** (Product C) and **financial impact** (Product W) rather than relying on a single metric.
